```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import sklearn
import datetime as dt

```

```python
df = pd.read_csv('kc_house_data.csv')
#df.describe
df.head(10)
# Requirement is to create a model to predict Price (Linear Regression Model)
```

```python
#Converting Date to Pandas Date Time format
df['date']  = pd.to_datetime(df['date']) #Can be used to convert to Panda Format Date Time i.e. yyyy-mm-dd
df.head(10)
df['day'] =  df['date'].dt.day
#df['month'] = df.date.dt.to_period('M')
df['month'] = df['date'].dt.month
df['year'] = df.date.dt.to_period('A')
df = df.drop('date',1).drop('id',1) #Drop date Column. Alsodrop ID Column as it is Unique
df.head(10)
```

```python
df['is_duplicate'] = df.duplicated(keep=False)
#print(df[df.is_duplicate=='False'])
```

```python
pd.isnull(df) #No NA or Nulls
```

```python
#df=df.drop('is_duplicate',1)
df['price'] = df['price'].astype(int)  #bathrooms floors lat long
df['bathrooms'] = df['bathrooms'].astype(int)  #bathroomsfloorslatlong
df['floors'] = df['floors'].astype(int)  #bathroomsfloorslatlong
df['lat'] = df['lat'].astype(int)  #bathroomsfloorslatlong
df['long'] = df['long'].astype(int)  #bathroomsfloorslatlong
df.dtypes
```

```python
train=df.sample(frac=0.75,random_state=200)
test=df.drop(train.index)
#test.head
```

```python
#Divide the train and test data into Independent and Dependent Variables 
#X is having all Features or Independent Variables
#y will have Price Dependent or predicting variable
X = train.iloc[:,1:]
y= train.iloc[:,:1]
#Similarly for Test
X_test = test.iloc[:,1:]
y_test = test.iloc[:,:1]
```

```python
from sklearn.linear_model import LinearRegression
# Representing LinearRegression as lr(Creating LinearRegression Object)
lm = LinearRegression()
```

```python
lm.fit(X,y)  #Linear Regression 
```
